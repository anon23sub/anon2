```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
class QuantileRegressionPyTorchMSE(nn.Module):
    def __init__(self, input_length):
        super(QuantileRegressionPyTorchMSE, self).__init__()
        self.input_length = input_length
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=30, kernel_size=10)
        self.conv2 = nn.Conv1d(in_channels=30, out_channels=40, kernel_size=6)
        self.dropout = nn.Dropout(0.2)
        self.fc1 = nn.Linear(40 * (self.input_length - 10 - 6 + 2), 1024)
        self.fc2 = nn.Linear(1024, 1)

    def forward(self, X, deterministic=False):
        X = F.relu(self.conv1(X))
        X = F.relu(self.conv2(X))
        X = self.dropout(X) if not deterministic else X
        X = torch.flatten(X, 1)
        X = F.relu(self.fc1(X))
        X = self.dropout(X) if not deterministic else X
        output = self.fc2(X)
        return output

def mse_loss(pred, y):
    return F.mse_loss(pred, y)
```

```python
input_length = 99
model_mse = QuantileRegressionPyTorchMSE(input_length=input_length).to("cuda")
optimizer = optim.Adam(model_mse.parameters(), lr=0.001)
mse_losses = []
```

```python
class QuantileRegressionPyTorchMulti(nn.Module):
    def __init__(self, alphas, input_length):
        super(QuantileRegressionPyTorchMulti, self).__init__()
        self.alphas = alphas
        self.input_length = input_length
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=30, kernel_size=10)
        self.conv2 = nn.Conv1d(in_channels=30, out_channels=40, kernel_size=6)
        self.dropout = nn.Dropout(0.2)
        self.fc1 = nn.Linear(40 * (self.input_length - 10 - 6 + 2), 1024)
        # Two separate output layers for each quantile
        self.fc2_1 = nn.Linear(1024, 1)
        self.fc2_2 = nn.Linear(1024, 1)

    def forward(self, X, deterministic=False):
        X = F.relu(self.conv1(X))
        X = F.relu(self.conv2(X))
        X = self.dropout(X) if not deterministic else X
        X = torch.flatten(X, 1)
        X = F.relu(self.fc1(X))
        X = self.dropout(X) if not deterministic else X
        quantile1 = self.fc2_1(X)
        return quantile1

    def loss_fn(self, pred, y, alpha):
        errors = y - pred
        loss = torch.max(alpha * errors, (alpha - 1) * errors)
        return torch.mean(loss)
```

```python
alpha = (1-0.9)/2
model = QuantileRegressionPyTorchMulti(alphas=alpha, input_length=input_length).to("cuda")
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Lists to store losses for plotting
losses_01 = []
# losses_09 = []

# Training loop for 3 epochs
num_epochs = 50
for epoch in range(num_epochs):
    optimizer.zero_grad()
    quantile1 = model(x_train)
    loss_01 = model.loss_fn(quantile1, y_train,alpha)
    total_loss = loss_01 
    total_loss.backward()
    optimizer.step()
    
    # Saving the losses for this epoch
    losses_01.append(loss_01.item())


# Plotting the losses over epochs
plt.figure(figsize=(10, 6))
plt.plot(losses_01, label='Quantile 0.1 Loss', marker='o')
plt.title('Quantile Regression Losses Over 3 Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.xticks(range(num_epochs))
plt.legend()
plt.grid(True)
plt.show(), (losses_01, losses_09)
```

```python
def ace(ideal, predicted):
    """
    dataframe : pandas dataframe with Ideal and Counts as column for regression calibration
    It can be directly used as 2nd output from calibration_regression in plot.py 
    """
    def rmse_loss(y, yhat):
        return jnp.abs(y-yhat)
    return jnp.mean(jax.vmap(rmse_loss, in_axes=(0, 0))(ideal, predicted))
```
