# Underconfident

```python
true_values = [185,   208,        175,        260,        210,
               178,        220,        255,        215,        205]
predicted_means = [185,   200,        195,        245,
                   190,        185,        205,        240,        195,        210]
intervals = [(170, 200), (185, 215), (180, 210), (220, 270), (168, 212),
             (170, 200), (185, 225), (220, 260), (170, 220), (195, 225)]

# Time points
time = np.arange(1, 11)

# plt.figure(figsize=(5, 2.6))
# Plotting the true values
plt.scatter(time, true_values, label='Ground Truth', color='blue', s=11)

# Plotting the predicted means
plt.plot(time, predicted_means, label='Prediction',
         color='red', linestyle='--')

# Plotting the confidence intervals
# plt.fill_between(time, [i[0] for i in intervals], [i[1] for i in intervals], color='red', alpha=0.3)
plt.fill_between(time, [i[0] for i in intervals], [
                 i[1] for i in intervals], color='red', alpha=0.3, label='Predicted CI')
# #ED7117
# Set x-axis ticks at every integer value
plt.xticks(np.arange(1, 11, 1))

# Adding labels and title
# plt.title('Underconfident Intervals')
# plt.xlabel('Time')
# plt.ylabel('Power (W)')
# plt.xticks()
# plt.yticks()
plt.title('Underconfident Intervals', fontsize=12)
plt.xlabel('Time\n(b)', fontsize=10)
plt.ylabel('Power (W)', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add grid background
plt.grid(True, alpha=0.2)

# set y limit
plt.ylim(150, 280)
plt.legend()
plt.legend(frameon=False)


# Show the plot

plt.savefig("/home/haikoo.khandor/haikoo/final conformal/underconfident.pdf")
```
